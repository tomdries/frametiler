```python
%load_ext autoreload
%autoreload 2
from frametiler import VideoSampler, Tiler
from IPython.display import Image

## Use the VideoSampler class to sample frames from a video
video_sampler = VideoSampler("test_files/example_video.mp4")
video_sampler.sample_fixed_n(n_frames=6, t_start=0, t_end=6)  # Sample fixed amount of frames
# video_sampler.sample_fixed_rate(sample_rate=1, t_start=0, t_end=6.8)  # Alternatively: Fixed sample rate in seconds
# video_sampler.save_frames_to_folder("frames_folder") # Optinal: Save frames to folder

## Use the Tiler class to create a mosaic from the sampled frames
tiler=Tiler(video_sampler.to_array())  
mosaic = tiler.create_mosaic(n_rows=2, border_size=2, annotation_style='number', annotation_position='topleft')
# mosaic = tiler.resize_mosaic(mosaic, 800, 600) # Optional: Resize mosaic to fit in a certain size
tiler.save_mosaic_to_image(mosaic, "test_files/mosaic.jpg")

Image(filename='test_files/mosaic.jpg')
```
